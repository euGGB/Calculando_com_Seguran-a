```python
import sys

# Constantes com os preços dos itens de segurança
#Itens e preços de cada um em inteiro
#aqui atribuímos os valores respectivos dos materiais de segurnaça
ITENS_SEGURANCA = {
    "alarme": 3000,
    "cameras": 1900,
    "sensores": 1600,
    "PIR_básicos": 100,
    "mov_inteligente": 200
}

#Def para apresentação da empresa
def apresentacao_empresa():
    """Exibe a mensagem de apresentação da empresa"""
    print("Olá! Nós somos a Calculando_com_Segurança")
    print("Nós estamos aqui para te ajudar a planejar a segurança da sua moradia...")
    print("\n...visando praticidade, objetividade com economia e qualidade.")
    print("===============================================================")

#
def mostrar_itens_disponiveis():
    """Exibe a lista de itens de segurança disponíveis"""
    print("\nAs medidas de segurança recomendadas para uma moradia incluem:")
    print(f"• Sistema de alarme R$ {ITENS_SEGURANCA['alarme']:.2f}")
    print(f"• Sistema completo de câmeras de segurança R$ {ITENS_SEGURANCA['cameras']:.2f}")
    print(f"• Kits com múltiplos sensores de movimento R$ {ITENS_SEGURANCA['sensores']:.2f}")
    print("\n===============================================================")

def obter_orcamento_cliente():
    """Obtém e valida o orçamento do cliente"""
    while True:
        preco_input = input("Com base em sua disponibilidade financeira, informe o valor que pretende investir nesses itens: ")
        preco_input = preco_input.replace(',', '.').replace(' ', '')
        so_digitos = ''.join(c for c in preco_input if c.isdigit() or c == '.')  # Limpa a entrada

        if so_digitos:
            return int(float(so_digitos))
        print("Por favor, insira um valor numérico válido.")

def verificar_orcamento_minimo(preco_cliente):
    """Verifica se o orçamento atende ao mínimo necessário"""
    while preco_cliente < 100.00:
        print("Poxa, seu orçamento está um pouco abaixo do esperado para os produtos recomendados...\n")
        atendimento = input("Deseja encerrar o atendimento? (sim/não)").strip().lower()

        if atendimento == "sim":
            print("Muito obrigado por confiar em nosso serviço! Até a próxima.")
            sys.exit()
        elif atendimento == "não":
            print("Vamos continuar! Qual vai ser o seu novo orçamento?")
            preco_cliente = obter_orcamento_cliente()
        else:
            print("Opção inválida, tente novamente.")

    return preco_cliente

def recomendar_seguranca(preco_cliente):
    """Faz recomendações com base no orçamento do cliente"""
    print("\n--- Recomendação de segurança com base no seu orçamento ---")

    if preco_cliente == 100:
        reais = 100
        print(f"\nCom base no valor informado, recomendamos os sensores de movimento PIR básicos, custo R$ {reais:.2f}")
    elif 200 <= preco_cliente < 1600:
        reais = 200
        print(f"\nCom base no valor informado, recomendamos os sensores de movimento inteligente, custo R$ {reais:.2f}")
    elif 1600 <= preco_cliente < 1900:
        reais = 1600
        print(f"\nCom base no valor informado, recomendamos os Kits com múltiplos sensores de movimento, custo R$ {reais:.2f}")
    elif 1900 <= preco_cliente < 3000:
        reais = 2100
        print(f"\nCom base no valor informado, recomendamos o sistema completo de câmeras de segurança e os sensores de movimento inteligente, custo R$ {reais:.2f}")
    elif 3000 <= preco_cliente < 4600:
        reais = 3000
        print(f"\nCom base no valor informado, recomendamos o sistema de alarme, custo R$ {reais:.2f}")
    elif 4600 <= preco_cliente < 4900:
        reais = 4600
        print(f"\nCom base no valor informado, recomendamos o sistema de alarme e os Kits com múltiplos sensores de movimento, custo R$ {reais:.2f}")
    elif 4900 <= preco_cliente < 6500:
        reais = 4900
        print(f"\nCom base no valor informado, recomendamos o sistema de alarme e o sistema completo de câmeras de segurança, custo R$ {reais:.2f}")
    elif 6500 <= preco_cliente:
        reais = 6500
        print("\nCom base no valor informado, recomendamos todas as medidas de segurança.")
    else:
        print("Resposta inválida.")
        return None

    print(f"{reais} é a quantidade olhe ofertada")
    return reais

def mostrar_opcoes_pagamento():
    """Exibe as opções de pagamento disponíveis"""
    print("Bem-vindo ao sistema de pagamento!")
    print("========================")
    print("      [1] - CRÉDITO       ")
    print("      [2] - DÉBITO        ")
    print("      [3] - PIX           ")
    print("========================\n")

def escolher_forma_pagamento():
    """Obtém a forma de pagamento escolhida pelo cliente"""
    return input("\nQual será sua forma de pagamento? ").strip()

def calcular_parcelas(reais):
    """Calcula e valida o parcelamento para pagamento no crédito"""
    while True:
        parcelas = input("Você deseja parcelar em quantas vezes? Nosso limite máximo para parcelamento é 3x: ").strip()

        if parcelas not in ["1", "2", "3"]:
            print("Nosso limite é parcelamento até 3X.\n")
            trocar = input("Deseja trocar a forma de pagamento? (sim/não): ").strip().lower()

            if trocar == "sim":
                return None
            elif trocar == "não":
                compra = input("Deseja continuar com a compra mesmo assim? (sim/não): ").strip().lower()
                if compra != "sim":
                    print("Muito obrigado por confiar em nosso serviço! Até a próxima.")
                    sys.exit()
            else:
                print("Opção inválida.")
        else:
            return int(parcelas)

def obter_valor(mensagem):
    """Obtém e valida um valor numérico do usuário"""
    while True:
        valor = input(mensagem).replace(",", ".")
        try:
            return float(valor)
        except ValueError:
            print("Valor inválido. Use apenas números.")

def processar_pagamento_singular(reais):
    """Processa pagamento usando apenas uma forma de pagamento"""
    forma_pagamento = escolher_forma_pagamento()

    if forma_pagamento == "1":
        print("Você selecionou CRÉDITO.\n")
        parcelas = calcular_parcelas(reais)
        if parcelas is not None:
            valor_parcela = reais / parcelas
            print(f"Você irá pagar {parcelas}x de R$ {valor_parcela:.2f}\n")
    elif forma_pagamento == "2":
        print(f"Você escolheu pagar à vista via DÉBITO. O valor é R$ {reais:.2f}")
    elif forma_pagamento == "3":
        print(f"Você escolheu pagar à vista via PIX. O valor é R$ {reais:.2f}")
    else:
        print("Opção inválida. Por favor, escolha 1, 2 ou 3.")

def processar_pagamento_opcoes(reais):
    """Processa pagamento dividido entre diferentes formas"""
    opcao = input("Quais das opções abaixo você deseja?\n"
                 "==================================\n"
                 "      [1] - CRÉDITO E DÉBITO      \n"
                 "      [2] - DÉBITO E PIX          \n"
                 "      [3] - PIX E CRÉDITO         \n"
                 "==================================\n").strip().lower()

    if opcao == "1":
        valor_credito = obter_valor("Quanto deseja pagar no crédito? ")
        parcelas = calcular_parcelas(valor_credito)
        valor_debito = reais - valor_credito
        print(f"Você escolheu parcelar R${valor_credito:.2f} em {parcelas}x e pagar R${valor_debito:.2f} à vista no débito.")
    elif opcao == "2":
        valor_debito = obter_valor("Quanto deseja pagar no débito? ")
        valor_pix = reais - valor_debito
        print(f"Você escolheu pagar R${valor_debito:.2f} no débito e R${valor_pix:.2f} no PIX.")
    elif opcao == "3":
        valor_credito = obter_valor("Quanto deseja pagar no crédito? ")
        parcelas = calcular_parcelas(valor_credito)
        valor_pix = reais - valor_credito
        print(f"Você escolheu parcelar R${valor_credito:.2f} em {parcelas}x e pagar R${valor_pix:.2f} no PIX.")
    else:
        print("Opção inválida. Por favor, escolha 1, 2 ou 3.")

def despedida():
    """Exibe mensagem de despedida"""
    print("=============================================================")
    print("Muito obrigado por confiar em nosso serviço! E até a próxima.")
    print("A calculando_com_segurança te espera para uma futura compra.")

def main():
    """Função principal que orquestra todo o fluxo do programa"""
    apresentacao_empresa()
    mostrar_itens_disponiveis()

    preco_cliente = obter_orcamento_cliente()
    preco_cliente = verificar_orcamento_minimo(preco_cliente)

    reais = recomendar_seguranca(preco_cliente)
    if reais is None:
        return

    mostrar_opcoes_pagamento()
    possibilidade_pag = input("\nDeseja pagar de forma singular ou dividir entre as outras opções de pagamento? (singular/opções)").strip().lower()

    if possibilidade_pag == "singular":
        processar_pagamento_singular(reais)
    elif possibilidade_pag == "opções":
        processar_pagamento_opcoes(reais)
    else:
        print("Opção inválida.")

    despedida()

if __name__ == "__main__":
    main()
```
